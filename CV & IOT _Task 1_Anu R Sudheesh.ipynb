# EasyOCR

**EasyOCR is a python package that allows the image to be converted to text.**

```python
Lang_reader = easyocr.Reader(['en'])
```

# Detection of texts from the image

```python
output1 = Lang_reader.readtext('F:\Internship\CV Task 1 images\HP1.jpg')
```

# Print the detected texts from image

```python
output1
```

# Extract the fetched output text & bounding box coordinates

```python
cord1 = output1[-2][0]
print(output1[-2][1])
```

# Minimum value of coordinates


```python
x_min1,y_min1 = [min(idx) for idx in zip(*cord1)]
```

# Maximum value of coordinates

```python
x_max1,y_max1 = [max(idx) for idx in zip(*cord1)]
```

# Visualization of bounding box co-ordinates of particular text


**Detecting the text "phials" in the image.**

```python
image = cv2.imread('F:\Internship\CV Task 1 images\HP1.jpg')
cv2.rectangle(image,(int(x_min1),int(y_min1)),(int(x_max1),int(y_max1)),(0,255,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
